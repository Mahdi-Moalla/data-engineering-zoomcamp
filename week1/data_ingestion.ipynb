```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
data_df=pd.read_parquet('./green_tripdata_2025-11.parquet')
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:postgres@localhost:5433/ny_taxi')
```

```python
print(pd.io.sql.get_schema(data_df, name='taxi_data', con=engine))
```

```python
data_df.to_sql(name='taxi_data', con=engine, if_exists='replace')
```
